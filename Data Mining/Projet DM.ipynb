# Visualisation de la distribution 

```python
from matplotlib.colors import ListedColormap

def plot_distribution(clf, X, y, axes=[-1.5, 2.45, -1, 1.5], alpha=0.5, contour=True):
    
    h = .02
    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    
    Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=.5)
    
    plt.scatter(X[:,0], X[:,1], c=y, alpha=0.8, edgecolors='k')
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, noise=0.3, random_state=0)
plt.scatter(X[:,0], X[:,1], c=y, alpha=0.8)
```

### Modeles Machine Learning

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42, stratify=y)
```

### 1/ RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
modele1 = RandomForestClassifier(n_estimators=100)
modele1.fit(X_train, y_train)
modele1.score(X_test, y_test)
```

```python
#Courbe ROC 
from sklearn.metrics import plot_roc_curve
plot_roc_curve(modele1, X_test, y_test)
plt.title("La Courbe ROC")
```

# Arbre de décision

```python
from sklearn.tree import DecisionTreeClassifier
modele3=DecisionTreeClassifier(max_depth=2, random_state=4)
modele3.fit(X_train, y_train)
print('L\'acuracy sur les données du test : {:.3f}'.format(modele3.score(X_test, y_test)))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(10,12))
plot_tree(modele3,feature_names = list(X),filled=True)
```

### SVM(SVC )

```python
from sklearn.svm import SVC
modele2 = SVC(kernel='linear')
#modele2 = SVC(kernel='rbf')
modele2.fit(X_train, y_train)
modele2.score(X_test,y_test)
```

```python
y_pred2 = modele2.predict(X_test)
```

```python
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
conf_matrix = confusion_matrix(y_test,y_pred2)
print(conf_matrix)
print(f1_score(y_test,y_pred2))
plot_confusion_matrix(modele2, X_test, y_test)
```
